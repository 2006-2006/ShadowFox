```python
#Line Plot
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.title("Line Plot")
plt.show()
```

```python
#Scatter Plot
import matplotlib.pyplot as plt

x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
#Bar Chart
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.title("Bar Chart")
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(170, 10, 250)

plt.hist(data)
plt.title("Histogram")
plt.show()
```

```python
#Pie Chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.title("Pie Chart")
plt.show()
```
